```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b-')
plt.show()

```

```python
data = {
    'Item': ['Espresso', 'Latte', 'Cappuccino', 'Mocha'],
    'Total_Revenue': [237.50, 262.50, 81.25, 67.50]
}
df_summary = pd.DataFrame(data)

df_summary = df_summary.set_index('Item')
```

```python
plt.bar(df_summary.index, df_summary['Total_Revenue'])
plt.show()
```

```python

time_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Subscribers': [10000, 12000, 15000, 14500, 18000]
}
df_subs = pd.DataFrame(time_data)
display(df_subs)
```

```python

genre_data = {
    'Genre': ['Action', 'Comedy', 'Drama', 'Sci-Fi'],
    'Watch_Hours': [450, 300, 200, 250]
}
df_genres = pd.DataFrame(genre_data).set_index('Genre')
display(df_genres)
```

```python
plt.plot(df_subs['Month'], df_subs['Subscribers'],'r-')
plt.show()
```

```python
plt.barh(df_genres.index, df_genres['Watch_Hours'])
plt.show()
```

```python
plt.pie(df_genres['Watch_Hours'],labels=df_genres.index)
plt.show()
```
